# Power Law Behavior

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from utils import data_fit, fit
```

## Data

```python
data_water = pd.read_csv('./data/5water.csv')
data_glyce = pd.read_csv('./data/5glyce.csv')

fps = 8810

dropoff_frame_water = 41
dropoff_frame_glyce = 78

frames_water = (dropoff_frame_water - data_water['frame'])
frames_glyce = (dropoff_frame_glyce - data_glyce['frame'])

t_water = frames_water/fps
t_glyce = frames_glyce/fps

t_err_water = 1/(2 * fps)
t_err_glyce = 1/(2 * fps)

r_water = data_water['width']/2
r_glyce = data_glyce['width']/2

r_err_water = data_water['width_err']/2
r_err_glyce = data_glyce['width_err']/2
```

# Water

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.errorbar(
    np.log(t_water), np.log(r_water), xerr=t_err_water/t_water, yerr=r_err_water/r_water, 
    color='red', alpha=0.5, fmt='.', 
    label='Water'
)

ax.set_xlabel(r'$\ln (t_d - t)$')
ax.set_ylabel(r'$\ln r$')
ax.grid(alpha=0.5)
ax.set_title('Drop pinch-off of Water')

plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
params, params_err, chisq, dof = fit(
    np.log(t_water), np.log(r_water), r_err_water/r_water, 
    0, 20, 
    r'$\ln (t_{d} - t)}$', 
    r'$\ln r$', 
    'Drop pinch-off of Water', 
    ax=ax
)
```

## 20/40 Water/Glycerine

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.errorbar(
    np.log(t_glyce), np.log(r_glyce), xerr=t_err_glyce/t_glyce, yerr=r_err_glyce/r_glyce, 
    color='red', alpha=0.5, fmt='.', 
    label='glycerine'
)

ax.set_xlabel(r'$\ln (t_d - t)$')
ax.set_ylabel(r'$\ln r$')
ax.grid(alpha=0.5)
ax.set_title('Pinch Drop-off of 20/40 Water-Glycerine Mix')

plt.tight_layout()
```
