# Top Pinch-Off

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from utils import *
```

## Data

```python
data_top = pd.read_csv('./data/6.csv')
data_bottom = pd.read_csv('./data/5water.csv')

fps = 8810

dropoff_frame_top = 35
dropoff_frame_bottom = 41

frames_top = (dropoff_frame_top - data_top['frame'])
frames_bottom = (dropoff_frame_bottom - data_bottom['frame'])

times_top = frames_top/fps
times_bottom = frames_bottom/fps

diameters_top = data_top['width']
diameters_bottom = data_bottom['width']

diameters_err_top = data_top['width_err']
diameters_err_bottom = data_bottom['width_err']
```

## Top and Bottom Neck Pinchoffs

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.errorbar(
    np.log(times_bottom), np.log(diameters_bottom), yerr=diameters_err_bottom/diameters_bottom, 
    color='blue', alpha=0.5, fmt='.', 
    label='Bottom'
)

ax.errorbar(
    np.log(times_top), np.log(diameters_top), yerr=diameters_err_top/diameters_top, 
    color='red', alpha=0.5, fmt='.', 
    label='Top'
)

ax.set_xlabel('Log-time till drop-off (log-seconds)')
ax.set_ylabel('Log-diameter of neck (log-pixels)')
ax.grid(alpha=0.5)
ax.set_title('Pinch Drop-off')
ax.legend()
```

## Top Pinch-off
