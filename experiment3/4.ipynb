```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from utils import *
```

## Data

```python
data = pd.read_csv('./data/4.csv')

fps = 15968
dropoff_frame = 1054
frames = (dropoff_frame - data['frame'])
times = frames/fps

diameters = data['width']/2

diameters_err = data['width_err']
```

## Checkpoint 4

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.errorbar(
    np.log(times), np.log(diameters), yerr=diameters_err/diameters, 
    color='blue', 
    alpha=0.5,
    fmt='.'
)

ax.set_xlabel('Log-time till drop-off (log-seconds)')
ax.set_ylabel('Log-diameter of neck (log-pixels)')
ax.grid(alpha=0.5)
ax.set_title('Pinch Drop-off')
```
