# Power Law Behavior

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from utils import data_fit, fit
```

## Data

```python
data_water = pd.read_csv('./data/5water.csv')
data_glyce = pd.read_csv('./data/5water.csv')

fps = 8810

dropoff_frame_water = 41
dropoff_frame_glyce = 78

frames_water = (dropoff_frame_water - data_water['frame'])
frames_glyce = (dropoff_frame_glyce - data_glyce['frame'])

times_water = frames_water/fps
times_glyce = frames_glyce/fps

diameters_water = data_water['width']
diameters_glyce = data_glyce['width']

diameters_err_water = data_water['width_err']
diameters_err_glyce = data_glyce['width_err']
```

# Water

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.errorbar(
    np.log(times_water), np.log(diameters_water), yerr=diameters_err_water/diameters_water, 
    color='red', alpha=0.5, fmt='.', 
    label='Water'
)

ax.set_xlabel('Log-time till drop-off (log-seconds)')
ax.set_ylabel('Log-diameter of neck (log-pixels)')
ax.grid(alpha=0.5)
ax.set_title('Drop pinch-off of Water')

plt.tight_layout()
```

```python
params, params_err, chisq, dof = fit(
    np.log(times_water), np.log(diameters_water), diameters_err_water/diameters_water, 
    0, 15, 
    r'$\log (t_{pinch} - t)}$', 
    r'$\log (D)$', 
    'Drop pinch-off of Water'
)
```

```python
params, params_err, chisq, dof = fit(
    np.log(times_water), np.log(diameters_water), diameters_err_water/diameters_water, 
    15, 20, 
    r'$\log (t_{pinch} - t)}$', 
    r'$\log (D)$', 
    'Drop pinch-off of Water'
)
```
