# Top Pinch-Off

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from utils import *
```

## Data

```python
data_top = pd.read_csv('./data/6.csv')
data_bottom = pd.read_csv('./data/5water.csv')

fps = 8810

dropoff_frame_top = 35
dropoff_frame_bottom = 41

frames_top = (dropoff_frame_top - data_top['frame'])
frames_bottom = (dropoff_frame_bottom - data_bottom['frame'])

t_top = frames_top/fps
t_bottom = frames_bottom/fps

r_top = data_top['width']/2
r_bottom = data_bottom['width']/2

r_err_top = data_top['width_err']/2
r_err_bottom = data_bottom['width_err']/2
```

## Top and Bottom Neck Pinchoffs

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.errorbar(
    np.log(t_bottom), np.log(r_bottom), yerr=r_err_bottom/r_bottom, 
    color='blue', alpha=0.5, fmt='.', 
    label='Bottom'
)

ax.errorbar(
    np.log(t_top), np.log(r_top), yerr=r_err_top/r_top, 
    color='red', alpha=0.5, fmt='.', 
    label='Top'
)

ax.set_xlabel(r'$\ln (t_d - t)$')
ax.set_ylabel(r'$\ln r$')
ax.grid(alpha=0.5)
ax.set_title('Top and Bottom Pinch Drop-off')
ax.legend()
```

## Top Pinch-off

```python
fig, ax = plt.subplots(figsize=(5, 5))
params, params_err, chisq, dof = fit(
    np.log(t_top), np.log(r_top), r_err_top/r_top, 
    0, 20, 
    r'$\ln (t_d - t)}$', 
    r'$\ln r$', 
    'Top Pinch-Off', 
    ax=ax
)
```
