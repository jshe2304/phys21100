# 4. Calibrating the horizontal field sweep

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from utils import *
```

```python
data = pd.read_csv('./data/voltages.csv')
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

params, params_err, chisq, dof = fit(
    data['scope'], data['meter'], data['meter_err'], 
    #fit_range=(0, 20), 
    ax=ax
)

ax.set_xlabel('$V_s$ (volts)')
ax.set_ylabel('$I_m$ (amps)')
ax.set_title('Calibrating oscilloscope voltage\n using metered current')

txt = '$I_m = a \\cdot V_s + b $\n'
txt += f'$a = {params[0]:.2f} \\pm {params_err[0]:.2f}$ \n'
txt += f'$b = {params[1]:.2f} \\pm {params_err[1]:.2f}$ \n'
txt += f'$\\chi^2 = {chisq:.2f}$ \n'
txt += f'DOF$ = {dof}$'
ax.text(0.02, 0.85, txt, transform=ax.transAxes , fontsize=8, va='top');
```
